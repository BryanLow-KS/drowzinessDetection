# 4. Real Time Detections



```python
cap = cv2.VideoCapture(0)

#############################################
# fourcc = cv2.VideoWriter_fourcc(*'DIVX')        
# resolution = (480, 270)
# filename = 'output.mp4'
# videoOutput = cv2.VideoWriter(filename, fourcc, 10, (480,270))   
#############################################


while cap.isOpened():
    ret,frame = cap.read()
    frame = cv2.resize(frame, (480, 270), interpolation = cv2.INTER_AREA)
    
    #make detections
    results = model(frame)
    
    cv2.imshow('YOLO', np.squeeze(results.render()))
    
#     videoOutput.write(frame)
    
    cv2.resizeWindow('YOLO', 480, 270)
    
    if cv2.waitKey(10) & 0xFF == ord('q'): 
        break

cap.release()
cv2.destroyAllWindows()
```

# 5. Train from scratch

```python
import uuid #unique identifier
import os
import time
```

```python
IMAGES_PATH = os.path.join('data', 'images') #/data/images
labels = ['awake', 'drowsy']
number_imgs = 20
```

```python
cap = cv2.VideoCapture(0)
# Loop through labels
for label in labels:
    print('Collecting images for {}'.format(label))
    time.sleep(5)
    
    # Loop through image range
    for img_num in range(number_imgs):
        print('Collecting images for {}, image number {}'.format(label, img_num))
        
        # Webcam feed
        ret, frame = cap.read()
        
        # Naming out image path
        imgname = os.path.join(IMAGES_PATH, label+'.'+str(uuid.uuid1())+'.jpg')
        
        # Writes out image to file 
        cv2.imwrite(imgname, frame)
        
        # Render to the screen
        cv2.imshow('Image Collection', frame)
        
        # 2 second delay between captures
        time.sleep(2)
        
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```

```python
import pygame
pygame.init()
pygame.mixer.music.load("warningAlarm.wav")
```

```python
cap = cv2.VideoCapture(0)
warningCount = 1

while cap.isOpened():
    ret,frame = cap.read()
    frame = cv2.resize(frame, (480, 270), interpolation = cv2.INTER_AREA)
        
    #make detections
    results = model(frame)
    
    if results.pred[0].shape == torch.Size([0,6]):
        warningCount = 1
    
    else:
        if int(results.pred[0][0][5]) == 1: 
#             print('wake up!', warningCount)
            warningCount += 1
            if warningCount > 55: 
                pygame.mixer.music.play()
        else: 
            warningCount = 1

        
    
    cv2.imshow('YOLO', np.squeeze(results.render()))
    
    cv2.resizeWindow('YOLO', 480, 270)
    
    if cv2.waitKey(10) & 0xFF == ord('q'): 
        break

cap.release()
cv2.destroyAllWindows()
```
